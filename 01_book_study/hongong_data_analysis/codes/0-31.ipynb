```python
count_df = ns_df[['도서명', '저자', 'ISBN', '권', '대출건수']]
loan_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()

#원본 데이터프레임 중복된 행 삭제하기 
ns_df2 = ns_df.drop_duplicates(subset=['도서명', '저자', 'ISBN', '권']).copy()
ns_df2.duplicated(subset=['도서명', '저자', 'ISBN', '권']).sum()

#원본 데이터프레임의 선택 열들을 인덱스로 만들기(update의 겹침 기준은 칼럼 키가 아닌 인덱스 번호/라벨이기 때문!)
ns_df2.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_df2

#원본 데이터프레임과 병합하기
ns_df2.update(loan_df)
ns_df2.reset_index(inplace=True)
ns_df2 = ns_df2[ns_df.columns]
```

```python
#일괄 처리 함수 만들기(필요 없는 행 및 열을 삭제 -> 중복된 행을 대출건수 기준으로 병합하는 함수)
import pandas as pd
import numpy as np

#함수 정의
def data_cleaning(csv_file):
    
    #파일을 데이터프레임으로 읽기
    dfO = pd.read_csv(csv_file, low_memory=False, encoding='utf-8-sig')
    df = dfO.copy()
    
    #결측치 제거
    df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
    df.dropna(axis=1, how='all', inplace=True)
    df.dropna(how='all', inplace=True)
    
    #대출건수를 도서명, 저자, ISBN, 권수 기준으로 병합하여 새로운 데이터 프레임 만들기
    count_df = df[['도서명', '저자', 'ISBN', '권', '대출건수']]
    loan_count_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()
    
    #기존 데이터프레임과 합치기
    df.drop_duplicates(subset=['도서명', '저자', 'ISBN', '권'], inplace=True)
    df.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
    
    df.update(loan_count_df)
    df = df.reset_index()
    
    
    return df

gdown.download(r'https://bit.ly/3RhoNho', 'ns_202104.csv', quiet=False)

```

```python

ns_new_book = data_cleaning('ns_202104.csv')
ns_new_book
```
