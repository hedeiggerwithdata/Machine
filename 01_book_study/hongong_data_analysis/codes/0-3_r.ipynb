```python
import gdown
import pandas as pd

#중복된 책 병합 + 대출건수 계산하여 함께 포함한 새로운 데이터프레임의 csv파일 만들기

gdown.download('https://bit.ly/3RhoNho', 'ns_202104.csv', quiet=False)
ns_df = pd.read_csv(r'ns_202104.csv', low_memory=False)
ns_book = ns_df.dropna(how='all', axis=1)
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()

ns_book3 = ns_book.drop_duplicates(subset=['도서명', '저자', 'ISBN', '권']).copy()
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_book3.update(loan_count)
ns_book4 = ns_book3.reset_index()
ns_book4 = ns_book4[ns_book.columns]

ns_book4.to_csv("ns_book4.csv", index=False)
```

```python
#도서 데이터 클린징(불필요 데이터 삭제 및 병합) 함수

def bookdata_dup(csv_file):
    data_df = pd.read_csv(csv_file, low_memory=False)
    dup_book = data_df.duplicated(subset = ['도서명', '저자', '권', 'ISBN'])
    return sum(dup_book)

def bookdata_cleansing(csv_file):
    data_df = pd.read_csv(csv_file, low_memory=False)
    orgn_df = data_df.dropna(how='all', axis=1)
    
    count_df = orgn_df[['도서명', '저자', '권', 'ISBN', '대출건수']]
    group_df = count_df.groupby(by=['도서명', '저자', '권', 'ISBN'], dropna=False).sum()
    
    book_df = orgn_df.drop_duplicates(['도서명', '저자', '권', 'ISBN']).copy()
    book_df.set_index(['도서명', '저자', '권', 'ISBN'], inplace=True)
    book_df.update(group_df)
    book_df = book_df.reset_index()
    
    book_df = book_df[orgn_df.columns]
    
    return book_df

bookdata_dup("ns_202104.csv")
new_book_df = bookdata_cleansing("ns_202104.csv")


print(new_book_df.equals(ns_book4))
print(new_book_df.equals(r"C:\data\ns_book4.csv"))
```
