```python
#기술통계 판다스 메서드
import re
import numpy as np
ns_book7.describe()
ns_book8 = ns_book7.copy()

s = ns_book8['주제분류번호'].astype('string').str.strip()
f = ~ s.str.contains(r'\d', regex=True, na=True)

def cleansing(x):
    global s
    if x == '<NA>':
        x = np.nan
    if pd.isna(x):
        return '-1'
    if str(x).strip() == '':
        return '-1'
    else:   
        return str(x).strip()[0]
    
주제분류번호 = ns_book8['주제분류번호'].apply(cleansing)
ns_book8['주제분류번호'] = 주제분류번호

ns_book8['주제분류번호'].mode()
```
