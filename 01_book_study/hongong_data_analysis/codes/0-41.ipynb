```python
import gdown
import pandas as pd
import matplotlib.pyplot as plt


gdown.download(r'https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
import numpy as np

rd = pd.to_datetime(ns_book7['등록일자'], format='%Y-%m-%d', errors='coerce')
st = pd.Timestamp('2022-12-31')
timegap = (st - rd).dt.days
ns_book7['등록일자'] = timegap
```

```python
ns_book7['실질대출건수'] = ns_book7['대출건수'] / timegap
```

```python
plt.rcParams['font.family'] = 'malgun gothic'
npub = ns_book7['발행년도'] <= 2100
ns_book7.plot.scatter('발행년도', '등록일자', alpha=0.1)
plt.show()
```
