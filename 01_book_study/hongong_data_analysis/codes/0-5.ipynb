```python
import gdown
import pandas as pd

gdown.download(r'https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
pb_ct = ns_book7['출판사'].value_counts().head(10)

# reyd = pd.to_datetime(ns_book7['등록일자']).dt.strftime('%Y-%m')
# ns_book7['등록일자'] = reyd

rg = ns_book7['등록일자'].value_counts().sort_index()
pb_ct
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'malgun gothic'

fig , axs = plt.subplots(2, 2, figsize=(15, 9))
axs[0, 0].hist(ns_book7['대출건수'], bins=50)
axs[0, 1].scatter(ns_book7['도서권수'], ns_book7['대출건수'])
axs[1, 0].bar(pb_ct.index, pb_ct.values, width=0.7)
axs[1, 0].set_xticks(pb_ct.index, fontsize=4)
axs[1, 1].plot(rg.index, rg.values)
axs[1, 1].set_xlabel('연월', fontsize=7)
axs[1, 1].tick_params(axis='x', labelrotation=45, labelsize=1)
axs[1, 1].set_yscale('log')



```

```python
cy = ns_book7['발행년도'].value_counts().sort_index()
cy = cy[cy.index<=2030]

def cl_no(x):
    if pd.isna(x):
        return '-1'
    s = str(x).strip()
    if s == '':
        return '-1'
    else:
        return s[0]

sn = ns_book7['주제분류번호'].apply(cl_no).value_counts().sort_values(ascending=False)
sn.drop(index='-1', inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'malgun gothic'

plt.plot(cy.index, cy.values, linestyle = '--', marker='.', color='red')
plt.title('발행년도 별 도서 개수')
plt.xlabel('발행년도', fontsize=8)
plt.xticks(range(1947, 2030, 10), fontsize=7)
plt.ylabel('도서개수')
for idx, val in cy[::10].items():
    plt.annotate(val, (idx, val), xytext=(0, 5), textcoords='offset pixels', ha='center') #ha는 수평조절, va는 수직조절
plt.show()

plt.bar(sn.index, sn.values, color='skyblue', width=0.7)
plt.title('주제분류 별 도서 개수', fontsize=10)
plt.xlabel('주제분류', fontsize=8)
plt.ylabel('도서개수', fontsize=8)
for idx, val in sn.items():
    plt.annotate(val, (idx, val), xytext=(0, 3), textcoords='offset pixels', ha='center', va='bottom', fontsize=7)
plt.show()
```
