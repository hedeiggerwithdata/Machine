```python
import gdown
import pandas as pd

gdown.download(r'https://bit.ly/3GisL6J', 'ns_book4.csv', quiet=False)

ns_book4 = pd.read_csv('ns_book4.csv', low_memory=False)
ns_book4.head()
```

```python
import numpy as np
ns_book4.loc[0, '도서권수'] = 1
ns_book4 = ns_book4.astype({'도서권수':'int64', '대출건수':'int64'})

ns_book4.loc[0, '부가기호'] = np.nan
ns_book4.head(2)
```

```python
ns_book4.loc[ns_book4['부가기호'].isna(), '부가기호'] = '없음'
ns_book4.fillna({'부가기호':'없음'}).isna().sum()
```

```python
# DataFrame의 <replace>와 <정규식>
# ns_book4.replace({'발행년도':{r'\d\d(\d\d)':r'\1'}}, regex=True)[100:102]

ns_book4.replace({'발행년도':{r"\d\d(\d\d)":r"\1"}, '저자':{r"(.*)\s\(지은이\)(.*)\s\(옮긴이\)":r"\1\2"}}, regex=True)
```

```python
invalid_number = ns_book4['발행년도'].str.contains('\D', na=True)
ns_book4[invalid_number]

ns_book5 = ns_book4.replace({'발행년도': {r'.*(\d{4}).*':r'\1'}}, regex=True)
invalid_number = ns_book5['발행년도'].str.contains('\D', na=True)
#print(invalid_number.sum()) 67개 
ns_book5[invalid_number]

ns_book5.loc[invalid_number, '발행년도'] = '-1'
ns_book5 = ns_book5.astype({'발행년도':'int64'})


#-----------------------이상값 교체하기----------------------------#
dk_year_row = ns_book5['발행년도'].gt(4000)
ns_book5.loc[dk_year_row, '발행년도'] = ns_book5.loc[dk_year_row, '발행년도'] - 2333

ns_book5['발행년도'].gt(4000).sum()
ns_book5.loc[dk_year_row, '발행년도'] = -1

old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
ns_book5.loc[old_books, '발행년도'] = -1

rows_minus = ns_book5['발행년도'].eq(-1)
rows_minus.sum()


```

```python
na_rows = ns_book5['저자'].isna() | ns_book5['출판사'].isna() \
    | ns_book5['도서명'].isna() | ns_book5['발행년도'].eq(-1)

ns_book5[na_rows]
```

```python
import re
import requests
import pandas
from bs4 import BeautifulSoup

def find_bookinfo(row):
    title = row['도서명']
    author = row['저자']
    pub = row['출판사']
    year = row['발행년도']
    
    #URL로 페이지 html 텍스트 가져오기
    url = r"https://www.yes24.com/product/search?domain=ALL&query={}"
    headers={"User-Agent":"Mozilla/5.0"}
    try:
        page_html = requests.get(url.format(row['ISBN']), headers=headers, timeout=3)
        page_html.raise_for_status()
    except requests.RequestException as err:
        return f"{row['ISBN']}: {err}"
    page_soup = BeautifulSoup(page_html.text, 'html.parser')
    
    #전체 태그 박스 찾기
    box_li = page_soup.select_one("li[data-goods-no][data-iy-no][data-statgb]")
    if box_li is None:
        return title, author, pub, year
    
    #제목 태그 찾기
    if pd.isna(title):
        title_gd = box_li.select_one("a.gd_name")
        title = title_gd.get_text('', strip=True) if title_gd else None
        
    #저자 태그 찾기
    if pd.isna(author):
        author_span = box_li.select_one("span.info_auth")
        author_list = author_span.select("a") if author_span else []
        author = ', '.join(a.get_text(strip=True) for a in author_list) if author_list else author
    
        
    #출판사 태그 찾기
    if pd.isna(pub):
        pub_span = box_li.select_one("span.authPub.info_pub")
        pub = pub_span.get_text(" ", strip=True) if pub_span else None
    
    #발행년도 태그 찾기
    if pd.isna(year):
        year_span = box_li.select_one("span.authPub.info_date")
        year_str = year_span.get_text(strip=True) if year_span else None
        year = re.findall(r'\d{4}', year_str)[0]
    
    return title, author, pub, year
```

```python
sample = ns_book5[na_rows].head(20).apply(find_bookinfo, axis=1, result_type='expand')
cols = ['도서명', '저자', '출판사', '발행년도']
sample.columns = cols
```

```python
gdown.download(r'https://bit.ly/3UJZiHw', 'ns_book5_update.csv', quiet=False)
ns_book5_update = pd.read_csv('ns_book5_update.csv', low_memory=False, index_col=0) #인덱스 칼럼의 열은 열의 0번째로 쓰겠다
ns_book5_update
```

```python
ns_book5.update(ns_book5_update)

na_rows = ns_book5['저자'].isna() | ns_book5['출판사'].isna() \
    | ns_book5['도서명'].isna() | ns_book5['발행년도'].eq(-1)
    
ns_book6 = ns_book5.dropna(subset=['저자', '출판사', '도서명'])
ns_book6 = ns_book6[ns_book6['발행년도'] != -1]
ns_book6.to_csv('ns_book6.csv', encoding='utf-8', index=False)
```
