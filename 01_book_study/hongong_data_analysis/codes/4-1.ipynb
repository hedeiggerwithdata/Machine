```python
import gdown
import pandas as pd

gdown.download(r'https://bit.ly/3736JW1', 'ns_book6.csv', quiet=False)
ns_book6 = pd.read_csv('ns_book6.csv', low_memory=False)
ns_book6.head()
ns_book6.astype({'발행년도':'int64'})
```

```python
ns_book6.head()
ns_book6.astype({'발행년도':'int64'})
```

```python
# 기술통계로 요약하기
import numpy as np
ns_book6.describe()

ns_book7 = ns_book6[ns_book6['도서권수'] > 0]

ns_book7.describe(percentiles=[0.3, 0.6, 0.9])
ns_book7.describe(include='object')
ns_book7.columns

ns_book7['대출건수'].mean() #11
ns_book7['대출건수'].median()#6

pd.Series([1, 2, 3, 4, 5]).quantile(0.9, interpolation='nearest')

borrow_10_flag = ns_book7['대출건수'] < 10
borrow_10_flag.mean()

ns_book7['대출건수'].quantile(0.65)

differ = ns_book7['대출건수'] - ns_book7['대출건수'].mean()
np.sqrt(np.sum(differ**2) / (differ.count()-1))

ns_book7.describe(include='object')

ns_book7 = ns_book7.astype({'발행년도':'int32'})
ns_book7['발행년도'].mode()
```

```python
import numpy as np



values, counts = np.unique(ns_book7['도서명'], return_counts=True)
idx_counts = np.argsort(-counts)
top_values = values[idx_counts]
top_counts = counts[idx_counts]

book_zip = zip(top_values[:10], top_counts[:10])

for values, counts in book_zip:
    print(values, int(counts))
    
    
a = [1, 10, 3, 6, 20]
df = pd.Series(a)

print(df.var(), df.std())
```

```python
#평균 대출건수가 가장 높은 10개 출판사 추출 명령어
loan_df = ns_book7.groupby(by=['출판사'], dropna=False)['대출건수'] \
        .mean() \
        .sort_values(ascending=False) \
        .to_frame('평균대출건수')
loan_df['순위'] = range(1, len(loan_df)+1)
loan_df = loan_df.reset_index().set_index('순위')
loan_df.head(10)
```

```python
#대출건수 25% ~ 75% 범위 내의 도서가 전체 도서 중 차지하는 비중
q1, q3 = ns_book7['대출건수'].quantile([0.25, 0.75])
q_array = (ns_book7['대출건수'] >= q1) & (ns_book7['대출건수'] <= q3)
q_rate = (q_array.sum() / len(ns_book7)) * 100

q_rate
```
