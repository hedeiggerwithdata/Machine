```python
import gdown
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 

gdown.download(r'https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
ns_book7.columns
book_count = ns_book7['도서명'].apply(len)
plt.scatter(book_count, ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
plt.hist(ns_book7['대출건수'], bins=1000)
plt.yscale('log')
plt.show()
```

```python
ns_book7['대출건수']
```

```python
plt.boxplot(ns_book7[['대출건수', '도서권수']], whis=(0,100))
plt.yscale('log')
plt.show()
```

```python
import matplotlib as mpl
ns_book = pd.read_csv('my_ns_book6.csv', encoding='utf-8', low_memory=False)
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
ns_book = ns_book[ns_book['발행년도']<=2026]
ns_book.plot.scatter('도서권수', '대출건수', alpha=0.1)
plt.show()
```
