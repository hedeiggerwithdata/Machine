```python
import pandas as pd

ns_book7 = pd.read_csv(r'C:\data\ns_book7.csv', low_memory=False)
ns_book7.columns

```

```python
from sklearn.model_selection import train_test_split
import numpy as np

#scikit0-learn 패키지의 train_test_split 메서드 : 랜덤하게 표본 데이터 추출하여 데이터 세트 나눔
train_set, test_set = train_test_split(ns_book7, test_size = 0.25, random_state = 42)

#사이킷 머신러닝은 입력 데이터를 2차원 배열(샘플 수 i+ 속성 수)로, 타깃을 1차원 시리즈 객체로 받음. + 입력은 대문자로, 타깃은 소문자로 표기하는 관례
X_train = train_set[['도서권수']]
y_train = train_set['대출건수']

#입력 데이터와 타깃 데이터를 이용해 선형 회귀 모델 훈련하기(만들기) LinearRegression 클래스
from sklearn.linear_model import LinearRegression
lr = LinearRegression() #선형회귀 객체 만들기,객체는 학습한 패턴을 저장하고 있음.
lr.fit(X_train, y_train) #선형회귀 훈련시키기 fit 메서드

#테스트 세트로 모델 평가하기, 결과: 결정계수
X_test = test_set[['도서권수']]
y_test = test_set['대출건수']

lr.score(X_test, y_test)

from sklearn.dummy import DummyRegressor
dumlr = DummyRegressor()
dumlr.fit(X_train, y_train)  
dumlr.score(X_test, y_test)

# #결정계수 1이면?
# lr.fit(y_train.to_frame(), y_train)
# lr.score(y_test.to_frame(), y_test)

# #선형회귀모델링: 입력과 타깃 데이터를 학습해 기울기와 절편을 도출 및 저장
# print(lr.coef_, lr.intercept_)
```

```python
#로지스틱 회귀 분석

#타깃 클래스를 0/1 | T/F로 변환 *사이킷런은 T/F를 자동으로 1/0으로 인식
br_mean = np.mean(ns_book7['대출건수'])
y_train_c = y_train > br_mean
y_test_c = y_test > br_mean

from sklearn.linear_model import LogisticRegression

logr = LogisticRegression()
logr.fit(X_train, y_train_c)
logr.score(X_test, y_test_c)

from sklearn.dummy import DummyClassifier
dum = DummyClassifier()
dum.fit(X_train, y_train_c)
dum.score(X_test, y_test_c)

logr = LogisticRegression()
logr.fit(X_train, y_train_c)
logr.score(X_test, y_test_c)
```
