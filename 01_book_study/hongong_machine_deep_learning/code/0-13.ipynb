```python
from sklearn.neighbors import KNeighborsClassifier

#분류분석, 이진분석, 클래스, 특성
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
target = [1]*35 + [0]*14

data = list([x, y] for x, y in zip(length, weight))

kn = KNeighborsClassifier()
kn.fit(data, target)
kn.score(data, target)
```

```python
import pandas as pd

fish_df = pd.DataFrame({'length': length, 'weight': weight})

kn2 = KNeighborsClassifier()
kn2.fit(fish_df, target)

for n in range(5, len(fish_df)+1):
    kn2.n_neighbors = n
    score = kn2.score(fish_df, target)
    if score < 1:
        print("한계치 미달\nk: %s, 정확도:%s" %(n, score))
        break

```
