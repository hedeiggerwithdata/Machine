```python
fish_data = np.column_stack((fish_length, fish_weight))
fish_target = np.concatenate((np.ones(35), np.zeros(14)))

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = \
    train_test_split(fish_data, fish_target, random_state=42, stratify=fish_target)
    

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.predict([[25, 150]])

dst, idx = kn.kneighbors([[25, 150]])

plt.scatter(train_input[:,0], train_input[:,1], alpha=0.2)
plt.scatter(train_input[idx, 0], train_input[idx, 1], marker='*', alpha=0.5)
plt.scatter(25, 150, marker='^', c='red')
plt.show()
```

```python
import numpy as np
from scipy import stats

m = np.mean(train_input, axis=0)
s = np.std(train_input, axis=0)

print(m)
print(s)

train_input_n = (train_input-m)/s
test_input_n = (test_input-m)/s
sample = ([[25, 150]] - m)/s

kn = KNeighborsClassifier()
kn.fit(train_input_n, train_target)
kn.score(test_input_n, test_target)
kn.predict(sample)
idx = kn.kneighbors(sample, return_distance=False)

plt.scatter(train_input_n[:, 0], train_input_n[:, 1], alpha=0.3)
plt.scatter(train_input_n[idx, 0], train_input_n[idx, 1], marker='*', alpha=0.6)
plt.scatter(sample[:, 0], sample[:, 1], marker='^', c='red')
plt.show()
```
