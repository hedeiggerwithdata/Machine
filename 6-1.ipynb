```python
#비지도학습

import numpy as np
import matplotlib.pyplot as plt

fruits = np.load(r'C:\Machine\fruits_300.npy')
print(fruits.shape)
print(fruits[0, 0, :])

plt.imshow(fruits[0], cmap='gray' )
plt.show()

plt.imshow(fruits[0], cmap='gray_r')
plt.show()

fig, axs = plt.subplots(1, 2, dpi=200)
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[205], cmap='gray_r')
```

```python
#2차원 배열 만들기: 계산 편의성

apple = fruits[:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)

plt.hist(apple.mean(axis=1), alpha=0.3, color='red', label='apple')
plt.hist(pineapple.mean(axis=1), alpha=0.3, color='brown', label='pineapple')
plt.hist(banana.mean(axis=1), alpha=0.3, color='yellow', label='banana')
plt.legend()
plt.show()

fig, axs = plt.subplots(1, 3, figsize=(20,5))
axs[0].bar(range(10000), apple.mean(axis=0))
axs[0].set_title('apple')
axs[1].bar(range(10000), pineapple.mean(axis=0))
axs[1].set_title('pineapple')
axs[2].bar(range(10000), banana.mean(axis=0))
axs[2].set_title('banana')
plt.show()
```

```python
apple_mean = apple.mean(axis=0).reshape(100, 100)
pine_mean = pineapple.mean(axis=0).reshape(100, 100)
ban_mean = banana.mean(axis=0).reshape(100, 100)

fig, axs = plt.subplots(1, 3, figsize=(20, 5), dpi=240)
axs[0].imshow(apple_mean, cmap='gray_r')
axs[1].imshow(pine_mean, cmap='gray_r')
axs[2].imshow(ban_mean, cmap='gray_r')
fig.show()

```

```python
import numpy as np

abs_dff = np.abs(fruits - apple_mean)
abs_mean = np.mean(abs_dff, axis=(1, 2))

apple_idx = np.argsort(abs_mean)[0:100]
apple_idx = apple_idx.reshape(10, 10)

fig, axs = plt.subplots(10, 10, figsize=(10, 10), dpi=200)
for n in range(10):
    for m in range(10):
        axs[n, m].imshow(fruits[apple_idx[n, m]], cmap='gray_r')
        axs[n, m].axis('off')
fig.show()
```

```python
ban_idx = np.argsort(np.mean(np.abs(fruits-ban_mean), axis=(1, 2)))[:100]
ban_idx = ban_idx.reshape(10, 10)

fig, axs = plt.subplots(10, 10, figsize=(10, 10), dpi=240)
for n in range(10):
    for m in range(10):
        axs[n, m].imshow(fruits[ban_idx[n, m]], cmap='gray_r')
        axs[n, m].axis('off')

```
